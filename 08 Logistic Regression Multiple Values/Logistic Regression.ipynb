**`binary classification`** : *the binary classification is used to predict binary operation yes or no* <br>
**`multivalued classification`** : *the multi valued classification is to predict the multiple classificatoin problem, ex to find the given image in 0 to 9 etc...,*

```python
# importing the modules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset
from sklearn.datasets import load_digits
```

```python
data = load_digits()
dir(data)
```

```python
# splitting x and y

x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

images in dataset

```python
for i in range(232, 235):
    print(f"\nThis image is {data.target[i]}\n")
    plt.imshow(data.images[i])    
    plt.show()
```

```python
# creating the model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
print(f"The actual 37th image is {y_test[37]}")
print(f"The predicted 37th image is {model.predict([x_test[37]])}")

print(f"\n\nThe actual 17th image is {y_test[17]}")
print(f"The predicted 17th image is {model.predict([x_test[17]])}")

print(f"\n\nThe actual 24th image is {y_test[24]}")
print(f"The predicted 24th image is {model.predict([x_test[24]])}")
```

### Accuracy Rate for my model

```python
model.score(x_test, y_test)
```

```python
y_predicted = model.predict(x_test)
confusion_matrix = confusion_matrix(y_test, y_predicted)
confusion_matrix
```

```python
plt.xlabel('predicted')
plt.ylabel('truth')
plt.title('confusion matrix')
sns.heatmap(confusion_matrix, annot=True)
plt.show()
```
