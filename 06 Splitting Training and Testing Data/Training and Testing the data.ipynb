#### splitting x and y

*by splitting x and y* we can get some random testing and training datasets, by training the model with occured training dataset, we can find out the score of the testing dataset, so we can met with **accuracy rate**

```python
# obtain x and y
x = data.drop(['Sell Price($)'], axis=1)
y = data['Sell Price($)']
```

```python
# splitting x and y dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head()
```

**note** : each and every time we call the training dataset the valuea could be changed, if not to change the values, use rnadom_state.

```python
# splitting x and y dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
x_train.head()
```

```python
# create the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

**score** : *it compares the predicted x_test values with actual y_test* values so than we can predict how the model was deviated from the results

```python
model.score(x_test, y_test)
```

our model is 87% accurate
